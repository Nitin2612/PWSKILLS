#Q1. What is the difference between a function and a method in Python?

##Function
- Definition: A block of reusable code that performs a specific task.
- Standalone: Functions are not tied to any particular object and can be called directly.
- Declaration: Defined using the def keyword.
- Usage: Can take arguments and return results.
##Method
- Definition: A function that is associated with an object.
-Bound to an Object: Methods are functions that operate on the data contained in the object they belong to.
- Called Using Objects: Accessed through the object using the dot (.) operator.
- Usage: Often used to interact with or modify the objectâ€™s data.

```python
#Function
def greet(name):
    return f"Hello, {name}!"

print(greet("Alice"))  # Output: Hello, Alice!
```

```python
#Method
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f"Hello, {self.name}!"

person = Person("Alice")
print(person.greet())  # Output: Hello, Alice!
```

#Q4. What is the purpose of the `return` statement in a Python function?
###The return statement in Python is used to exit a function and send a value back to the caller (the place where the function was called). It allows a function to produce a result that can be stored in a variable, used in calculations, or printed directly.

###Key Purposes of return
- Return a Value: It allows the function to send a computed result back to the caller.
- End Function Execution: It stops the execution of the function as soon as it is encountered.
- Return Multiple Values: It can return multiple values using tuples.
- Return Without a Value: If no value is specified, it returns None by default.

```python
#Returning a Single Value
def add(a, b):
    return a + b
result = add(3, 4)
print(result)
```

#Q5. What are iterators in Python and how do they differ from iterables?
###Iterable
An iterable is any Python object that can be looped over (iterated) using a for loop. It implements the __iter__() method, which returns an iterator object. Essentially, an iterable is a collection of items that can be traversed.
- Examples: Lists, tuples, dictionaries, sets, strings, etc.
- Common Use: Used in for loops to get the values one by one.
##Iterator
An iterator is an object that represents a stream of data. It allows you to iterate over the elements of an iterable one by one, keeping track of the current position. An iterator must implement the __iter__() and __next__() methods.
-  __iter__(): Returns the iterator object itself.
- __next__(): Returns the next item in the sequence. When there are no more items left, it raises the StopIteration exception to signal the end of iteration.



```python
# Example of an iterable
my_list = [1, 2, 3, 4]

# Convert iterable to an iterator
iterator = iter(my_list)

# Access elements using the `next()` function
print(next(iterator))  # 1
print(next(iterator))  # 2
print(next(iterator))  # 3
```

#Q6. Explain the concept of generators in Python and how they are defined.
###Generators in Python are a type of iterable, like lists or tuples, but they are more memory efficient. They allow you to iterate over large sequences of data without storing the entire sequence in memory. This is especially useful when working with large datasets or streams of data.
##How Generators Are Defined
Generators can be defined using:
- Generator Functions: These are functions that use the yield keyword to produce a series of values.
- Generator Expressions: Similar to list comprehensions but with parentheses instead of square brackets.


```python
#Generator Functions
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1

# Create a generator object
counter = count_up_to(5)

# Iterate over the generator using a for loop
for num in counter:
    print(num)
```

```python
# Generator expression to generate squares of numbers
squares = (x**2 for x in range(1, 6))

# Iterate over the generator
for square in squares:
    print(square)
```

#Q7. What are the advantages of using generators over regular functions?
###Advantages of Using Generators Over Regular Functions:
- Memory Efficiency: Generators yield items one at a time, so they do not store the entire sequence in memory, making them more memory-efficient for large datasets.
- Lazy Evaluation: Values are generated only when needed, reducing unnecessary computation and improving performance for large or infinite data sets.
- State Preservation: Generators maintain their state between iterations, allowing them to resume from where they left off without recalculating previous values.
- Convenience for Iteration: They provide a simple way to iterate over data without needing to manually manage indices or states.

```python
#Regular Function (List):
def create_numbers(n):
    numbers = []
    for i in range(1, n+1):
        numbers.append(i)
    return numbers

result = create_numbers(5)  # Creates a list of 5 numbers
print("Regular function result:", result)
```

```python
#Generator Function:
def generate_numbers(n):
    for i in range(1, n+1):
        yield i

result = generate_numbers(5)  # Returns a generator
print("Generator function result:")
for num in result:
    print(num)
```

#Q8. What is a lambda function in Python and when is it typically used?
###A lambda function in Python is a small, anonymous function defined using the lambda keyword. It can take any number of arguments but can only have one expression. The result of the expression is automatically returned without needing a return statement.

####Syntax of a Lambda Function:

- Short and Simple Functions: Lambda functions are ideal when you need a small, throwaway function for a short period, typically used in cases where defining a full function with def would be overkill.
- Higher-Order Functions: Lambda functions are commonly used in higher-order functions like map(), filter(), and sorted(), where a simple function is passed as an argument.



```python
# Lambda function that adds two numbers
add = lambda x, y: x + y
result = add(3, 5)
print(result)
```

#Q9. Explain the purpose and usage of the `map()` function in Python.
###The map() function in Python is a built-in function that allows you to apply a given function to all items in an iterable (such as a list, tuple, or string) and return an iterator that produces the results. The purpose of map() is to transform data in an iterable by applying a function to each element in the iterable.

####Syntax of map() function:
map(function, iterable)

##Use Cases for map():
- Transforming data in a list (e.g., squaring numbers, converting strings to uppercase).
- Applying the same operation across all elements in an iterable without using a loop.
- Combining multiple iterables into a single output based on some operation.

```python
#Example of Using map() with a Regular Function
def square(x):
    return x ** 2
numbers = [1, 2, 3, 4, 5]
squared_numbers = map(square, numbers)
print(list(squared_numbers))
```

```python
#Example of Using map() with a Lambda Function:
numbers = [1, 2, 3, 4, 5]
squared_numbers = map(lambda x: x ** 2, numbers)
print(list(squared_numbers))
```

#Q10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
###The map(), reduce(), and filter() functions in Python are all higher-order functions from the functools module (except map() which is built-in) and are commonly used for functional programming tasks.
1. map():
- Purpose: The map() function is used to apply a given function to every item in an iterable (like a list, tuple, etc.) and return an iterator that yields the results.
- Usage: It takes a function and one or more iterables, applies the function to each element of the iterable(s), and returns an iterator of the results.
####Syntax:

map(function, iterable)

map(function, iterable1, iterable2, ...)
- Key Concept: Applies a function to each element in the iterable(s) and returns the transformed data.
2. reduce():
- Purpose: The reduce() function from the functools module is used to apply a binary function (a function that takes two arguments) cumulatively to the items of an iterable, so as to reduce the iterable to a single value.
- Usage: It repeatedly applies the function to the elements of the iterable in a cumulative manner, starting from the first two elements.
###Syntax:

from functools import reduce

reduce(function, iterable)
- Key Concept: Reduces the iterable to a single value by applying the function cumulatively.
3. filter():
- Purpose: The filter() function is used to filter out elements from an iterable based on a given condition (i.e., a function that returns True or False).
- Usage: It takes a function (condition) and an iterable and returns an iterator of elements for which the function returns True.
####Syntax:
filter(function, iterable)
- Key Concept: Filters the elements of an iterable based on a condition and returns a subset of the iterable.

```python
#map():
numbers = [1, 2, 3, 4]
result = map(lambda x: x ** 2, numbers)
print(list(result))

#reduce():
from functools import reduce
numbers = [1, 2, 3, 4]
result = reduce(lambda x, y: x + y, numbers)
print(result)

#filter():
numbers = [1, 2, 3, 4, 5]
result = filter(lambda x: x % 2 == 0, numbers)
print(list(result))
```

#Q1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

```python
def sum_of_evens(numbers):
    return sum(num for num in numbers if num % 2 == 0)

user_input = input("Enter a list of numbers (as digits or space-separated numbers): ")

if ' ' in user_input:
    numbers = [int(num) for num in user_input.split()]
else:
    numbers = [int(digit) for digit in user_input]

result = sum_of_evens(numbers)
print(f"The sum of even numbers in the list is: {result}")
```

#Q4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(number):
    if number <= 1:
        return False
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            return False
    return True

number = int(input("Enter a number between 1 and 200: "))
if 1 <= number <= 200:
    if is_prime(number):
        print(f"{number} is a prime number.")
    else:
        print(f"{number} is not a prime number.")
else:
    print("Please enter a number between 1 and 200.")
```

#Q5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms
        self.current = 0
        self.next_num = 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.current
        elif self.count == 1:
            self.count += 1
            return self.next_num
        else:
            fib = self.current + self.next_num
            self.current = self.next_num
            self.next_num = fib
            self.count += 1
            return fib

terms = int(input("Enter the number of terms for the Fibonacci sequence: "))
fib_iterator = FibonacciIterator(terms)

for num in fib_iterator:
    print(num)
```

#Q6. Write a generator function in Python that yields the powers of 2 up to a given exponent.


```python
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

exponent = int(input("Enter the exponent: "))
for power in powers_of_two(exponent):
    print(power)
```

#Q7.  Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_line_by_line(file_name):
    with open(file_name, 'r') as file:
        for line in file:
            yield line.strip()

file_name = input("Enter the file name (with path if needed): ")
try:
    for line in read_file_line_by_line(file_name):
        print(line)
except FileNotFoundError:
    print("File not found. Please check the file path and try again.")
```

#Q9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

celsius_temps = [0, 20, 30, 40, 100]

fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

print(f"Temperatures in Fahrenheit: {fahrenheit_temps}")
```

#Q10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def is_not_vowel(char):
    return char.lower() not in 'aeiou'

input_string = input("Enter a string: ")

filtered_string = ''.join(filter(is_not_vowel, input_string))

print(f"String without vowels: {filtered_string}")
```
